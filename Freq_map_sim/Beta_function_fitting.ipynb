## Beta function fitting

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import mpld3
from pylab import *
from scipy import stats

#mpld3.enable_notebook()
```

Import data first

```python
data1 = np.genfromtxt('beta_fit')
# Remove NaN in data and change dtype to float
for i,clm in enumerate(data1.T):
	for j,tmp in enumerate(clm):
		if tmp =='NaN':
			data1[j,i] = np.nan
		else:
			data1[j,i] = float64(tmp)  
```

```python
slope, inter, r, p, std = stats.linregress(data1[:,2],data1[:,0]**2)
print slope, inter, r, p, std
```

```python
figure(2)
plt.plot(data1[:,2],marker='.',color='r')
plt.plot(data1[:,0]**2/slope,marker='.',linestyle='--',color='b')

plt.title('Beta function fit using linear amplitude fitting')
plt.xlabel('BPM #')
plt.ylabel('Beta X (m)')
plt.legend(('Model','Exp'))
plt.grid(True)
plt.show()
```

```python
figure(3)
plt.plot(np.abs(data1[:,2] - data1[:,0]**2/slope)/data1[:,2],marker='.',color='b')


plt.title('Beta beat of experimental with model')
plt.xlabel('BPM #')
plt.ylabel('$\Delta B / B$ (m)')

plt.grid(True)
plt.show()
```
