```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import mpld3
from pylab import *
from tuneDiagram import tuneDiagram

mpld3.enable_notebook()
```

```python
data1 = np.genfromtxt('FFT_mixed.txt',dtype=np.str)
data2 = np.genfromtxt('NAFF_mixed.txt',dtype=np.str)
data3 = np.genfromtxt('FFT_stacked.txt',dtype=np.str)
data4 = np.genfromtxt('NAFF_stacked.txt',dtype=np.str)
data5 = np.genfromtxt('TBT_BPM21.txt')
```

```python
figure(5)
plt.subplot(2,1,1)
plt.plot(data5[:,0])
plt.grid(True)
plt.title('Single BPM TBT data')
plt.ylabel('Horizontal (mm)')
plt.subplot(2,1,2)
plt.plot(data5[:,1])
plt.grid(True)
plt.ylabel('Vertical (mm)')
plt.xlabel('Turns #')
plt.show()
```

```python
# Remove NaN in data and change dtype to float
for i,clm in enumerate(data1.T):
	for j,tmp in enumerate(clm):
		if tmp =='NaN':
			data1[j,i] = np.nan
		else:
			data1[j,i] = float(tmp)
for i,clm in enumerate(data2.T):
	for j,tmp in enumerate(clm):
		if tmp =='NaN':
			data2[j,i] = np.nan
		else:
			data2[j,i] = float(tmp)
for i,clm in enumerate(data3.T):
	for j,tmp in enumerate(clm):
		if tmp =='NaN':
			data3[j,i] = np.nan
		else:
			data3[j,i] = float(tmp)
for i,clm in enumerate(data4.T):
	for j,tmp in enumerate(clm):
		if tmp =='NaN':
			data4[j,i] = np.nan
		else:
			data4[j,i] = float(tmp)
```

```python
figure(1,figsize=(12,5))
plt.subplot(2,1,1)
#plt.plot(data1[:,2],data1[:,0],marker='.',color='b')
plt.plot(data2[:,2],data2[:,0],marker='o',linestyle='-',color='b')
plt.grid(True)
plt.ylabel('Qx')
plt.title('Mixed BPM tune measurement')
#plt.legend('TEST')
plt.axis((0, len(data2[1:30,2]), 0.15,0.18))

plt.subplot(2,1,2)
#plt.plot(data1[:,2],data1[:,1],marker='.',color='b')
plt.plot(data2[:,2],data2[:,1],marker='o',linestyle='-',color='b')
plt.grid(True)
plt.xlabel('# turns')
plt.ylabel('Qy')
plt.axis((0, len(data2[1:30,2]), 0.21,0.24))
plt.show()
```

```python
figure(2,figsize=(12,5))
plt.subplot(2,1,1)
plt.plot(data3[:,2],data3[:,0],marker='.',color='b')
plt.plot(data4[:,2],data4[:,0],marker='.',linestyle='--',color='r')
plt.grid(True)
plt.ylabel('Qx tune')
plt.title('Stacked BPM horizontal tune measurement')
plt.legend(('NAFF'))
plt.axis((0, len(data2[:,2]), 0.15,0.18))

plt.subplot(2,1,2)
plt.plot(data3[:,2],data3[:,1],marker='.',color='b')
plt.plot(data4[:,2],data4[:,1],marker='.',linestyle='--',color='r')
plt.grid(True)
plt.xlabel('# turns')
plt.axis((0, len(data2[:,2]), 0.21,0.24))
```
